## Factorial

```python
Ns = [1, 5, 10, 100, 500, 1000]
Tn = [(5 * n) + 3 for n in Ns]

%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(Ns, Tn)
plt.ylabel('tiempo')
plt.xlabel('n')
plt.savefig('../plots/alg_fact_Tn.png')
plt.show()
```

```python
Ns = [1, 5, 10, 100, 200, 400, 500, 600, 800, 1000]
T1n = [(5 * n) + 3 for n in Ns]
T2n = [(10 * n) + 5 for n in Ns]

%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(Ns, T1n, '', label="$T_{1}(n)=5n+3$")
plt.plot(Ns, T2n, '-.', label="$T_{2}(n)=10n+5$")
plt.ylabel('tiempo')
plt.xlabel('n')
plt.legend()
plt.savefig('../plots/T1n_T2n.png')
plt.show()
```

```python
Ns = [1, 5, 10, 100, 200, 400, 500, 600, 800, 1000]
T1n = [(5 * n) + 3 for n in Ns]
T2n = [(10 * n) + 5 for n in Ns]
T3n = [(n ** 2) for n in Ns]

%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(Ns, T1n, '', label="$T_{1}(n)=5n+3$")
plt.plot(Ns, T2n, '-.', label="$T_{2}(n)=10n+5$")
plt.plot(Ns, T3n, ':', label="$T_{3}(n)=n^2$")
plt.ylabel('tiempo')
plt.xlabel('n')
plt.legend()
plt.savefig('../plots/T1n_T2n_T3n.png')
plt.show()
```

```python
Ns = [1, 5, 10, 100, 200, 400, 500, 600, 800, 1000]
T2n = [(10 * n) + 5 for n in Ns]
T3n = [(n ** 2) for n in Ns]
T4n = [(n ** 3) for n in Ns]

%matplotlib inline
import matplotlib.pyplot as plt

plt.plot(Ns, T2n, '-.', label="$T_{2}(n)=10n+5$")
plt.plot(Ns, T3n, '', label="$T_{3}(n)=n^2$")
plt.plot(Ns, T4n, ':', label="$T_{4}(n)=n^3$")
plt.ylabel('tiempo')
plt.xlabel('n')
plt.legend()
plt.savefig('../plots/T2n_T3n_T4n.png')
plt.show()
```

# Cotas

```python
from math import log, sin

Ns = range(1, 12)
O = [(4 * log(n)) for n in Ns]
Omega = [log(n) for n in Ns]
Tn = [(3 * log(n) + sin(n) - 0.15) for n in Ns]

plt.rc('text', usetex=True)
plt.plot(Ns, O, '-.', label="$c_{1} \cdot g(n)$")
plt.plot(Ns, Omega, ':', label="$c_{2} \cdot g(n)$")
plt.plot(Ns, Tn, label="$f(n) = \Theta(g(n))$")
plt.plot((2, 2), (0, 9), 'k--', label="$n_{0}$")
plt.ylabel('tiempo')
plt.xlabel('n')
plt.legend()
plt.savefig('../plots/cotas.png')
plt.show()
```

```python
Ns = range(1, 15)
Tn = [5 * n + 3 for n in Ns]
O = [6 * n for n in Ns]
Omega = [5 * n for n in Ns]

plt.rc('text', usetex=True)
plt.plot(Ns, O, '-.', label="$6n$")
plt.plot(Ns, Omega, ':', label="$5n$")
plt.plot(Ns, Tn, label="$T(n) = 5n + 3$")
plt.plot((3, 3), (0, 80), 'k--', label="$n_{0} = 3$")
plt.ylabel('tiempo')
plt.xlabel('n')
plt.legend()
plt.savefig('../plots/cotas_fact.png')
plt.show()
```

```python
Ns = range(0, 6)
Tn = [n ** 2 + 2 * n - 5 for n in Ns]
O = [2 * n ** 2 for n in Ns]
Omega = [n ** 2 for n in Ns]

plt.rc('text', usetex=True)
plt.plot(Ns, O, '-.', label="$2n^2$")
plt.plot(Ns, Omega, ':', label="$n^2$")
plt.plot(Ns, Tn, label="$T(n) = n^2 + 2n - 5$")
plt.plot((2.5, 2.5), (0, 50), 'k--', label="$n_{0} = 5/2$")
plt.ylabel('tiempo')
plt.xlabel('n')
plt.legend()
plt.savefig('../plots/cotas_selection_sort.png')
plt.show()
```

```python
from math import factorial

Ns = range(1, 150)
O1 = [1 for n in Ns]
Ologn = [10 * log(n) for n in Ns]
On = [2 * n for n in Ns]
Onlogn = [2 * n * log(n) for n in Ns]
On2 = [n ** 2 for n in Ns]
On3 = [n ** 3 for n in Ns]
O2n = [2 ** n for n in Ns]
Onfact = [factorial(n) for n in Ns]

plt.rc('text', usetex=True)
axes = plt.gca()
axes.set_ylim(ymax=1000)
plt.plot(Ns, O1, label="$1$")
plt.plot(Ns, Ologn, label="$\log n$")
plt.plot(Ns, On, label="$n$")
plt.plot(Ns, Onlogn, label="$n \log n$")
plt.plot(Ns, On2, label="$n^2$")
plt.plot(Ns, On3, label="$n^3$")
plt.plot(Ns, O2n, label="$2^n$")
plt.plot(Ns, Onfact, label="$n!$")
plt.ylabel('tiempo')
plt.xlabel('n')
plt.legend()
plt.savefig('../plots/jerarquia.png')
plt.show()
```
